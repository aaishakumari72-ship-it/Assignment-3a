Q1. What is data cleaning, and why is it important in data analysis?

>>Data cleaning is the process of finding and fixing errors in a dataset, such as missing values, duplicates, incorrect entries, or inconsistent formatting. It is important because dirty or inaccurate data leads to wrong results and misleading conclusions. Clean data ensures that your analysis is reliable, accurate, and useful for making good decisions.

 * What are the potential consequences of analyzing unclean or messy data?

 >>Analyzing unclean or messy data can lead to **wrong results**, **bad decisions**, and **misleading conclusions**. It can also cause **errors in models**, **waste time**, and make your analysis **unreliable**. In short, messy data = wrong outcomes.

 * Explain the common steps involved in cleaning and organizing data.
 >> 1. Remove Duplicates: Get rid of repeated data.
    2. Handle Missing Values: Fill or remove gaps in data.
    3. Fix Errors/Typos: Correct mistakes in data entries.
    4. Standardize Formats: Make data consistent (e.g., dates, names).
    5. Remove Irrelevant Data: Drop data you don't need.
    6. Validate Data: Check if data makes sense (e.g., age > 0).

Q3. xplain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

TRIM: Removes extra spaces from text.
 >>Ex: " Hello " → "Hello"

LEFT: Takes a certain number of characters from the start of a text.
>>Ex: LEFT("India123", 5) → "India"

RIGHT: Takes characters from the end of a text.
>>Ex: RIGHT("India123", 3) → "123"

MID: Extracts text from the middle of a string.
>>Ex: MID("ABC123XYZ", 4, 3) → "123"

CONCAT: Joins two or more pieces of text together.
>>Ex: CONCAT("Hello", " ", "World") → "Hello World"

Q8. Understanding TODAY() and NOW()

a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use
 each function.

>> TODAY() - It gives only the current date.Time not included.

   Ex: To calculate age or number of days left for an event.



NOW() - It gives current date AND time.

   Ex: To record the exact time a report was created or updated.


 b. If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are left until that date

 >>=TODAY()-A1

 c. Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY HH:MM AM/PM. ́

 >> =TEXT(NOW(), "MM/DD/YYYY hh:mm AM/PM")

 d. If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain
